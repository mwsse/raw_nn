## Derivatives

#### 01 - The impact of a parameter on the Output

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 2*x

x = np.array(range(5))
y = f(x)

print(x,y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
print((y[1]-y[0]) / (x[1] - x[0]))
```

```python
def f(x): 
    return 2*x**2

x = np.array(np.arange(-1, 5, 0.1))
y = f(x)

#print(x,y)
plt.plot(x,y)
plt.show()
```

```python
p2_delta = 0.0001

x1 = 1
x2 = x1 + p2_delta

y1 = f(x1)
y2 = f(x2)

approximate_derivative = (y2-y1)/(x2-x1)
print(approximate_derivative)
```
